```python
df_merged = df_energy.merge(
    df_geo,
    left_on="Entity",
    right_on="Country",
    how="left"
)

df_merged = df_merged.drop(columns=["Entity"])

print(df_merged.columns)
```

```python
df_merged = df_merged.drop(columns=["geometry"])
df_merged["geometry"] = df_merged["geometry_wkt"].apply(wkt.loads)
df_merged = df_merged.drop(columns=["geometry_wkt"])
```

```python
gdf = gpd.GeoDataFrame(df_merged, geometry="geometry")
```

```python
def plot_energy_transition(df, metric="Renewable_ratio", top_n=10, start_year=2000, end_year=2020):
    filtered = df[(df["Year"] >= start_year) & (df["Year"] <= end_year)]
    
    latest = filtered[filtered["Year"] == end_year]
    top_countries = latest.nlargest(top_n, metric)["Country"]
    
    filtered = filtered[filtered["Country"].isin(top_countries)]
    
    fig = px.line(
        filtered,
        x="Year",
        y=metric,
        color="Country",
        title=f"Top {top_n} Countries: {metric} Over Time"
    )
    
    fig.show()
```

```python
plot_energy_transition(
    gdf,
    metric="Renewable_ratio",
    top_n=8,
    start_year=2000,
    end_year=2010
)
```

```python
def plot_geo_energy(df, year=2020, metric="Energy_access_score"):
    filtered = df[df["Year"] == year]
    
    fig = px.choropleth(
        filtered,
        locations="Country",
        locationmode="country names",
        color=metric,
        hover_name="Country",
        color_continuous_scale="Viridis",
        title=f"{metric} by Country ({year})"
    )
    
    fig.show()
```

```python
plot_geo_energy(gdf)
```

```python
def plot_energy_mix(df, country="India", start_year=2000, end_year=2020):
    filtered = df[
        (df["Country"] == country) &
        (df["Year"] >= start_year) &
        (df["Year"] <= end_year)
    ]
    
    fig = px.area(
        filtered,
        x="Year",
        y=["Fossil_ratio", "Renewable_ratio", "Nuclear_ratio"],
        title=f"Energy Mix Evolution: {country}"
    )
    
    fig.show()
```

```python
plot_energy_mix(gdf)
```

```python
def plot_growth_vs_sustainability(df, year=2020, continent=None, degree=2):
    filtered = df[df["Year"] == year].copy()
    
    if continent:
        filtered = filtered[filtered["Continent"] == continent]
    
    filtered = filtered[(filtered["gdp_per_capita"] > 0) & (filtered["Renewable_ratio"].notna())]
    
    fig = px.scatter(
        filtered,
        x="gdp_per_capita",
        y="Renewable_ratio",
        size="Total_electricity",
        color="Continent",
        hover_name="Country",
        log_x=True,
        title=f"Growth vs Sustainability ({year})"
    )
    
    x = filtered["gdp_per_capita"]
    y = filtered["Renewable_ratio"]
    
    coeffs = np.polyfit(np.log10(x), y, degree)  
    poly = np.poly1d(coeffs)
    
    x_range = np.linspace(x.min(), x.max(), 200)
    y_pred = poly(np.log10(x_range))
    
    fig.add_trace(
        go.Scatter(
            x=x_range,
            y=y_pred,
            mode="lines",
            name=f"Polynomial Fit (deg={degree})",
            line=dict(width=3)
        )
    )
    
    fig.show()
```

```python
plot_growth_vs_sustainability(gdf)
```
